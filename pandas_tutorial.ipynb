```python
# apply

sony = ["PS", "PS2", "PS3", "PS4", "PSP", "PSV"]
nintendo = ["N64", "GB", "GBA", "GC", "Wii", "WiiU", "DS", "NES", "3DS", "SHES"]
microsoft = ["XB", "XB360", "XOne"]
pc = ["PC"]
atari = ["2600"]
panasonic = ["3DO"]
sega = ["DC", "GG", "SAT", "SCD"]
```

```python
def categories(x):
    if x in sony:
        return "Sony"
    if x in nintendo:
        return "Nintendo"
    if x in microsoft:
        return "Microsoft"
    if x in pc:
        return "PC"
    if x in atari:
        return "Atari"
    if x in panasonic:
        return "Panasonic"
    if x in sega:
        return "Sega"
    else:
        return "Other"
```

```python
df["Platform"].apply(categories).value_counts()
```

```python
df["Platform_Company"] = df["Platform"].apply(categories)
```

```python
# corr
corr_matrix = df.select_dtypes("float").corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, mask=np.triu(corr_matrix))
```

### Concatenating `DataFrames` with `pd.concat()`

```python
men60 = pd.read_csv("https://raw.githubusercontent.com/cajjster/data_files/main/men1960s.csv")
women60 = pd.read_csv("https://raw.githubusercontent.com/cajjster/data_files/main/women1960s.csv")
```

```python
men60.head(2), women60.head(2)
```

```python
pd.concat(objs=[men60, women60], ignore_index=True)
```

```python
pd.concat(objs=[men60, women60], keys=["Men 60's", "Women 60's"])
```

```python
pd.concat(objs=[men60, women60], axis=1)
```

```python
men = [f"https://raw.githubusercontent.com/cajjster/data_files/main/men{y}s.csv" for y in range(1920, 2021, 10)]
women = [f"https://raw.githubusercontent.com/cajjster/data_files/main/women{y}s.csv" for y in range(1920, 2021, 10)]
```

```python
pd.concat([pd.read_csv(file) for file in men], keys=list(range(1920, 2021, 10)))
```

```python
swe_women = pd.concat([pd.read_csv(file) for file in women], keys=list(range(1920, 2021, 10)))
```

```python
swe_women.groupby("Name")["Count"].sum().sort_values(ascending=False)
```

### Joining `DataFrames` with `merge()`

```python
boys19 = pd.read_csv("https://raw.githubusercontent.com/cajjster/data_files/main/boys2019.csv")
boys20 = pd.read_csv("https://raw.githubusercontent.com/cajjster/data_files/main/boys2020.csv")
```

```python
pd.options.display.max_rows = 500
```

```python
boys19.merge(boys20, how="outer", on="Name", suffixes=["_2019", "_2020"])
```

```python
boys19.merge(boys20, how="inner", on="Name", suffixes=["_2019", "_2020"])
```

```python
boys19.merge(boys20, how="right", on="Name", suffixes=["_2019", "_2020"])
```

```python
boys19.merge(boys20, how="left", on="Name", suffixes=["_2019", "_2020"])
```

```python
boys19_20 = boys19.merge(boys20, how="outer", on="Name", suffixes=["_2019", "_2020"], indicator=True)
boys19_20._merge.value_counts()
```
